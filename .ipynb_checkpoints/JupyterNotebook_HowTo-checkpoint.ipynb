### <center> General Interface
___
    
The materials included in PHYS 2331 are written in Markdown, and the code will generally be in python. 

Running each cell in the notebook will display the output regardless of whether it is text- or code-based.
    
#### <center> Binder
    
Binder is used to share a Jupyter repository from GitHub.

Once the link is opened, the main directory for the repository will display.

### <center> Interacting with the Jupyter Notebook
___
    
* Each box is called a cell. Double-clicking into the cell will reveal editable material in Markdown or Python (in this course). 

* The **Run** button will run the cells (or Shift + Enter).

    * In a Markdown cell, the material will appear within the same cell with a white background.
    
    * In a code cell, clicking Run will run that cell and the output will apear below. 
    
        * Note that code can be split in a Jupyter Notebook so that it can run in separate chunks. Verify that the correct run order is completed to limit errors.
        
    * If an unexpected error occurs, there may be an issue with the kernel (ex. timeout). If this occurs, click **Kernel > Restart and Clear Output** in the gray toolbar at the top of the notebook. The cells can then be run again per the user's disgression.
