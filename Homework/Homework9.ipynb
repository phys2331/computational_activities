```python
import scipy.constants as sp
import math

# Given variables. CHANGE THESE TO MATCH YOURS!

height = 62
length = 17
current = 3*10**4
freq = 60
h_body = 2
w_body = 0.5

# Pre-Calculations

r = math.sqrt(height**2 + ((length**2)/4))
sin_theta = length/(2*r)
ang_freq = 2 * sp.pi * freq
area_body = h_body * w_body

# Calculate

B = (((sp.mu_0*current)/(2*sp.pi*r))*2)*sin_theta
max_emf = B * area_body * ang_freq
max_emf_mV = max_emf * 1000

# Print answer
             
print("Part 1: The magnitude of the magnetic field =", B, "T")
print("Part 2: The maximum emf =", max_emf_mV, "mV")
```

### Important information

* Need:

* Given information:

#### Running the code

Change the following variables in the code box below to match your question in Wiley:

   * dp

   * mag_cent
   
   * loc
   
Now press the Run button! The answers should output below the box.

```python
import scipy.constants as sp
import math
from numpy import array

# Given variables. CHANGE THESE TO MATCH YOURS!

dp = array([0,0,8])

v = array([0,0,8])

mag_cent = array([1,4,4])

loc = array([1.02,4,3])

# Pre-Calculations

time = 1*10**-15
area_flux = sp.pi*(loc[0]-mag_cent[0])**2
vt = v * time
cent2 = vt + array([mag_cent[0],mag_cent[1],dp[2]])
mu = sp.mu_0/(4*sp.pi)
r1 = loc - mag_cent
r2 = loc - cent2
rmag1 = math.sqrt(r1[0]**2 + r1[1]**2 + r1[2]**2)
rmag2 = math.sqrt(r2[0]**2 + r2[1]**2 + r2[2]**2)
B_initial = ((mu*2)/(rmag1**3))*dp
B_final = ((mu*2)/(rmag2**3))*dp
phi_initial = B_initial * area_flux
phi_final = B_final * area_flux
delta_phi = phi_final - phi_initial

# Calculate

E = delta_phi/(time * sp.pi * 2 * r1[0])
E_field = array([0,E[2],0])

# Print answer
             
print("The electric field =", E_field ,"N/C")
```
