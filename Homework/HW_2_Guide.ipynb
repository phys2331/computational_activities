### <center> Important Information
    
**Objective:** Find the electric field at a given location due to a given charge, find the net electric field at a given location, and find the net force on a given location.

**Given Information:**

* Charges of the three particles ($Q_{1}, Q_{2}, Q_{3}$).

* Coordinates: +x is to the right, +y is up the page, +z is out of the page.

* Each grid square is 1 cm x 1 cm.

**Key Equations:**

\begin{equation*}
\vec{r} = observed - source
\end{equation*}

\begin{equation*}
\frac{1}{4\pi\epsilon_{0}} \frac{q}{|\vec{r}|^{2}} \hat{r}
\end{equation*}

\begin{equation*}
\vec{E_{net}} = \vec{E_{1}} + \vec{E_{2}} + \vec{E_{3}} ...
\end{equation*}

\begin{equation*}
\vec{F} = q\vec{E}
\end{equation*}
