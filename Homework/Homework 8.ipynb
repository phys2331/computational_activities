```python
import scipy.constants as sp
import math

# Given variables. CHANGE THESE TO MATCH YOURS!

r_disk = 0.02

E_field = 312
    
angle_init = 26

# Pre-Calculations

area_disk = sp.pi * (r_disk**2)
theta = 90-angle_init
   
# Calculate

E_flux = E_field * area_disk * math.cos(math.radians(theta))

# Print answer

print("The electric flux =", E_flux, "Vm")
```

![](Graphics/P21.07.PNG "Problem 21.07 - Matter and Interactions")


### Important information

* Need: 

    * Electric flux on each face of the box.
    
    * Total flux.
    
    * Total amount of change inside the box.

* Given information:

    * Height of box (h_box)
    
        * Change to meters in code.
    
    * Width of box (w_box)
    
        * Change to meters in code.
        
    * Length of box (l_box)
    
        * Change to meters in code.
        
    * $E_{1}$ going *into* the box (E_1)
    
    * $E_{2}$ going *out* of the box (E_2)
    
* Key equations:

\begin{equation*}
\phi_{E} = \int \vec{E} \bullet d\vec{A} = \Sigma \vec{E} \bullet \hat{n} \Delta A = |E||A|cos(\theta)
\end{equation*}

\begin{equation*}
\Sigma \vec{E} \bullet \hat{n} \Delta A = \frac{kQ}{r^{2}} = \frac{Q} {\epsilon_{0}}
\end{equation*}

<center> so 

\begin{equation*}
\phi_{E} \times \epsilon_{0} = Q
\end{equation*}

#### Running the code

Change the following variables in the code box below to match your question in Wiley:

   * h_box
   
   * w_box
   
   * l_box
   
   * E_1
   
   * E_2
   
Now press the Run button! The answers should output below the box.

```python
import scipy.constants as sp
import math

# Given variables. CHANGE THESE TO MATCH YOURS!

h_box = 0.03

w_box = 0.02
    
l_box = 0.08

E_1 = 340

E_2 = 1400

# Pre-Calculations

area_longsides = h_box * l_box
area_smallsides = h_box * w_box

theta_longsides = 0
theta_left = math.cos(math.radians(180))
theta_right = math.cos(math.radians(0))

epsnought = 8.85 * 10**-12
   
# Calculate

flux_longsides = E_1 * area_longsides * theta_longsides
flux_left = E_1 * area_smallsides * theta_left
flux_right = E_2 * area_smallsides * theta_right
flux_total = flux_longsides + flux_left + flux_right
charge = flux_total * epsnought

# Print answer

print("The flux through the top =", flux_longsides, "Vm")
print("The flux through the bottom =", flux_longsides, "Vm")
print("The flux through the left =", flux_left, "Vm")
print("The flux through the right =", flux_right, "Vm")
print("The flux through the front =", flux_longsides, "Vm")
print("The flux through the back =", flux_longsides, "Vm")
print("The total flux =", flux_total, "Vm")
print("The total charge in the box =", charge, "C")
```
