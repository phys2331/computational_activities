### <center> Curly Electric Fields
-----------
    
* **Constant current**: $B_{1}$ is constant in time and the electric field and magnetic fields **outside** the solenoid will be *essentially* **zero**. 


* **Varied current**: there is a **curly electric field** both outside and inside the solenoid. 

    * The magnetic field still does not exist outside of the solenoid with a varying current, as shown in **Figure 1**.

![](Art/CurlyEField.png "Curly Electric Field Produced by Changing Current")
**<center> Figure 1. A current varying over time in a solenoid produces a curly electric field.**
    
    
* **Inside the solenoid**: the electric field is *proportional* to r (distance from the axis). 

    * This means that as the point of interest nears the axis (decreasing r), the electric field is also decreasing. 

* **Outside the solenoid**: the curly electric field is proportional to $\frac{1}{r}$. 


* In contrast with the inside electric field, the curly electric field on the outside gets *smaller* as the point of interest moves away from the solenoid.


![](Art/DirectionCurlyEField.png "Magnetic field decreasing and increasing")
**<center> Figure 2. Four varying magnetic fields and rate of change.**


* **<u>Right-hand rule** Point the thumb of your right hand in the direction of the rate of change of the magnetic field ($\frac{-d\vec{B_{1}}}{dt}$) and curl fingers toward your palm. The direction in which the tip of the fingers are pointing is the direction of the curly electric field. 
    
    
* Find the direction of the rate of change of the magnetic field: change $\frac{-d\vec{B_{1}}}{dt}$ to $\frac{-\Delta \vec{B_{1}}}{\Delta t}$, where $\Delta t$ is a small, finite time interval. Then, the direction of $\frac{-\Delta \vec{B_{1}}}{\Delta t}$ = the direction of $-\Delta \vec{B_{1}}$.


* A solenoid with a ring over it will produce a current in the ring. 

    * The resulting $\vec{E_{NC}}$ will be in the same direction, as shown in **Figure 3**. Relate this to a battery, wire, and current.
    
![](Art/MetalRingSolenoid.png "The Effect of a Solenoid on a Metal Ring")
**<center> Figure 3. The Effect of a Solenoid on a Metal Ring**
        
* In **Figure 3**, the $I_{2}$ in the ring is proportional to the $\vec{E_{NC}}$ inside the metal. 

* To drive current in a wire, the wire must encircle a region where the magnetic field is changing. If the wire does not, there will be no emf present.

### <center> Faraday's Law
-----------
    
Faraday's Law explains the relationship between $\frac{-\Delta \vec{B_{1}}}{\Delta t}$ and $|\vec{E_{NC}}|$.

#### <center> Observing Current Caused by Non-Coulomb Electric Fields
    
* In order to measure the current in a loop around a solenoid, such as in Figure 4, an ammeter must be used. 

* If $I_{1}$ is changing in the solenoid, then $B_{1}$ will also change, causing $I_{2}$ to occur in the outer wire surrounding the solenoid.

#### <center> Magnetic Flux
    
* **Magnitude of the emf**: numerically equal to $|emf| = |\frac{d}{dt}(B_{1}\pi r^{2}_{1})|$


* **Magnetic flux**: calculated in the same way as the electric flux in Chapter 21 on Gauss's law.

    * $(B_{1}\pi r^{2}_{1})$ is the **magnetic flux $\phi_{mag}$** on the area encircled by the circuit.
       
    * On a small area: $\delta A$ is $\vec{B}\bullet\hat{n}\delta A = B_{\bot}\delta A$
    
        * $\hat{n}$: dimensionless unit vector perpendicular to the area.
        
            * "Normal" to the surface.

        * $B_{\bot}$: the perpendicular component of the magnetic field.
 
    * All contributions over the surface are summed to obtain the magnetic flux over the surface.
    
    * Can be *positive*, *negative*, or *zero*. 
    
        * Depends on the orientation of the magnetic field ($\vec{B}$) relative to $\hat{n}$. 

#### <center> Faraday's Law Quantitative
    
* **Faraday's Law**: in the presence of a time-varying magnetic field, there is an emf whose magnitude  is equal to the rate of change of magnetic flux.

    * Major physical law concerning the relationship between electric and magnetic fields. 
    
    * $emf = -\frac{d \phi_{mag}}{dt} \equiv\oint \vec{E_{NC}} \bullet \vec{dl} = -\frac{d}{dt} \int \vec{B}\bullet\hat{n} dA$

    * emf along a round-trip path = rate of change of the magnetic flux on the area encircled by the path. 
   
    * Direction: Calculate the magnetude of the effect. Then give direction (sign) based on the right-hand rule.     

* **Bigger emf**: Faster rate of change of magnetic flux.

    * Also occurs with a bigger area in the same magnetic field.  
    
#### <center> Coulomb Electric Field in Faraday's Law
        
* **Net Electric Field**: $\vec{E} = \vec{E_{C}} + \vec{E_{NC}}$

    * Remember: the round trip integral of a Coulomb electric field is **0** .
        
        * Therefore: 
        
\begin{equation*}        
emf = \oint \vec{E} \bullet \vec{dl} = \oint (\vec{E_{C}} + \vec{E_{NC}}) \bullet \vec{dl} = 0 + \oint \vec{E_{NC}} \bullet \vec{dl}
\end{equation*}
        
 
#### <center> Voltmeter Readings
    
* **Voltmeter**: acts like an ammeter with a large resistance in series.

* **Voltmeter leads**: when surrounding region of a changing magnetic field, the voltmeter will read an emf equal to $\frac{d\phi_{mag}}{dt}$. 
    
    * If the leads do not surround the changing magnetic field, the voltmeter will read **0 V**. See Figure for an example.    
    
#### <center> The EMF for a coil with Multiple Turns
    
* The emf in a coil containing *N* turns is *N* times the emf for one loop of the coil.

* If the loops are tightly wound so that they are very close together, $\vec{E_{NC}}$ is about the same in each loop.
   
#### <center> Faraday's Law and Moving Coils or Magnets
    
* Moving a coil with a steady current closer to a second coil **increases** the magnetic field and magnetic flux in the second coil.

    * Because the magnetic field is increasing, the emf can drive current in the second coil.
    
* Other ways to find an emf:

    * Moving a bar magnet  toward or away from a coil.
    
    * Rotating the bar magnet or coil
    
#### <center> Complications
    
* The $I_{2}$ in the second coil (from the first) also makes a magnetic field ($B_{2}$).

    * If $I_{2}$ is constant, $B_{2}$ will be constant, and will not have an additional $E_{NC}$.
    
    * If there is a time-varying rate of change, then $I_{2}$ is not constant and will have a time-varying magnetic field, $E_{NC}$, and emf of its own.
    
        * In most cases, this effect is small enough to ignore.

### <center> Faraday's Law and Motional EMF
    
--------
   
#### <center> Two Pieces of the Flux Derivative
    
* **Motional emf associated with magnetic forces**: if the magnetic field is constant but the shape or orientation of the loop changes to enclose more or less flux, the flux changes with time.

* Moving a coil toward a stationary magnet causes no change in magnetic field, but there is motional emf in the coil.
    
#### <center> Non-Coulomb Fields and Forces
    
* Effects that can be seen simultaneously:

    * Forces from Coulomb electric fields.
    
    * Forces from non-Coulomb electric fields from time-varying magnetic fields.
    
    * Magnetic forces ($q\vec{v}x\vec{B}$) if the wire is moving (producing motional emf).

### <center> Superconductors
    
--------

* **Superconductors**: materials that lose all resistance to electric current.

    * Below a critial temperature ($T_{C}$, the resistance diminishes completely.
    
        * Mobility becomes infinite.
        
        * A nonzero electric field in a superconductor would produce an infinite current.
    
#### <center> Magnetic Flux through a Superconducting Ring
    
* The magnetic flux cannot change through a superconducting ring.

    * Changing the flux through the ring by amount $\Delta\phi_{mag}$, produces a noninfinite current that produces an amount of flux $-\Delta\phi_{mag}$, resulting in a net flux change of 0. 
    
#### <center> The Meissner Effect
    
* **The Meissner Effect**: current runs in a solid superconducting disk in such a way that the net magnetic field is zero inside of the disk.

    * Applies to Type I superconductors.
    
    * Superconducting currents can run only on the surface of the superconductor, but not the interior.
    
#### <center> Magnetic Levitation
--------

* Meissner effect causes a magnet to hover abot a superconducting disk.

    * Current runs in the superconducting disk that repels that magnet, no matter its orientation.
    
        * Example: Maglev trains
    
### <center> Inductance
--------

* Changing the current in a coil leads to an *additional* emf in the coil itself.

    * Due to the coil surrounding a region of time-varying magnetic field produced *by itself*.
    
    * Acts in a direction to oppose the change in the current.
    
    * Causes a difficulty in changing the current (increase or decrease).
    
* The self-induced emf is proportional to the rate of change of the current. 

    * L = Proportionality constant.
    
        * Inductance (or, self-inductance) of the device (the inductor).

\begin{equation*}
|emf_{ind}| = L|\frac{dl}{dt}|
\end{equation*}

* The inductance of a solenoid is $L = \frac{\mu_{0} N^{2}}{d} \pi R^{2}$.

    * Units: the henry (H) $V \bullet s/A$

#### <center> Transformers
    
* In a transformer that has a primary coil with many turns that makes an emf in a second coil with a smaller amount of turns can convert high voltage to a lower voltage.

* The ratio of the number of turns determines the change in voltage.

* Small voltage in the second coil = larger current.

    * Due to energy conservation and $I\Delta V$.
    
* Step-down transformer: reduces the voltage and increases the current. Primary coil has many turns. Secondary coil has a few turns.

* Step-up transformer: increases the voltage and decreases the current. Primary coil has a few turns and the secondary coil has many turns.
    
#### <center> Energy Density
    
* There is energy density associated with magnetic fields.

* Magnetic energy stored in an inductor is $\frac{1}{2} LI^{2}$.

* Integration using power, magnetic energy, B, and L yields the electric and magnetic energy density equation:

\begin{equation*}
\frac{energy}{volume} = \frac{1}{2}\epsilon_{0}E^{2} + \frac{1}{2}\frac{1}{\mu_{0}}B^{2}
\end{equation*}

* Units: $J/m^{3}$

* Where there are electric or magnetic fields in space, there is an associated energy density, proportional to the square of the field ($E^{2}$ and/or $B^{2}$).

### <center> Inductor Circuits
--------

#### <center> Current in an RL Circuit
    
* Resistor (R) + inductor (L) = RL circuit.

* Current varies with time

\begin{equation*}
I = \frac{emf_{battery}}{R}(1 - e^{-(\frac{R}{L})t})
\end{equation*}

* An inductor in a circuit causes the current to take longer to reach the desired value than if it was excluded.

* Spark may occur across an opening switch because the time interval is short, $\frac{dI}{dt}$ is very large, and emf is very large.

    * Note: Don't open the switch in an inductive circuit if surrounded by explosive gases.

#### <center> Current in an LC Circuit
    
* Circuits may oscillate (not achieve a final equilibrium)

    * Resistance dampens oscillation.
    
* Inductor (L) + Capacitor (C) = LC circuit

    * Will oscillate if resistance is small.
    
* If capacitor runs out of charge, the current in the inductor will recharge the capacitor.

    * Due to inability of the current in the inductor to go to zero instantaneously.
    
    * Will oscillate until resistance is applied.
    
#### <center> Energy in an LC Circuit
    
* Electric energy in capacitor = $\frac{Q^{2}}{2C}$.

\begin{equation*}
Q = Q_{i} cos(\frac{1}{\sqrt{LC}}t)
\end{equation*}

* When Q = 0 in the capacitor, the current is max, and all the energy is the magnetic energy of the inductor.

* Resonance phenomenon

    * Applying AC voltage to force oscillations = small current unless free oscillation frequency ($f = \frac{1}{(2\pi\sqrt{LC})}$) is matched.
    
#### <center> AC Circuits
    
* When an AC current passes through an inductor, there's an AC emf generated due to the time-varying flux.

* If current = sine function, emf = cosine function.
    
#### <center> Skin Effect
    
* High-frequency phenomenon where the higher the AC frequency of the current, the bigger $\frac{d\phi}{dt}$.
    
    * $E_{2}$ << $E_{1}$ at very high frequencies.
    
    * Significant electric field only exists very near the surface of the wire.
    
    * Almost no current density in the wire due to $J = \frac{I}{A} = \sigma E$.

### <center> The Differential Form of Faraday's Law
--------

* Differential form of Faraday's law has a partial derivative with respect to time, when holding position constant.

\begin{equation*}
curl(\vec{E}) = -\frac{\partial\vec{B}}{\partial t}
\end{equation*}

<center> or

\begin{equation*}
\vec{\nabla} \times \vec{E} = -\frac{\partial\vec{B}}{\partial t}
\end{equation*}
    
--------    
### <center> Lenz's Rule
--------

* Can be used to get the direction of $E_{NC}$.

* $E_{NC}$ drives current in a direction to make a magnetic field that attempts to keep the flux constant.

* If $B_{1}$ is increasing at a constant rate, $I_{2}$ and magnetic field in a given ring are constant (Faraday's Law), so net flux *increases*.

* Lenz's rule also gives the direction of current for the following situations:
        
     * Motional emf.
     
     * Time-varying magnetic field and motion contribute to a change in flux
