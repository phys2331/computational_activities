### <center> Curly Electric Fields
-----------
    
* **Constant current**: $B_{1}$ is constant in time and the electric field and magnetic fields **outside** the solenoid will be *essentially* **zero**. 


* **Varied current**: there is a **curly electric field** both outside and inside the solenoid. 


    * The magnetic field still does not exist outside of the solenoid with a varying current, as shown in **Figure 1**.
    

![](Art/CurlyEField.png "Curly Electric Field Produced by Changing Current")
**<center> Figure 1. A current varying over time in a solenoid produces a curly electric field.**
    
    
* **Inside the solenoid**: the electric field is *proportional* to r (distance from the axis). 


    * This means that as the point of interest nears the axis (decreasing r), the electric field is also decreasing. 
    

* **Outside the solenoid**: the curly electric field is proportional to $\frac{1}{r}$. 


* In contrast with the inside electric field, the curly electric field on the outside gets *smaller* as the point of interest moves away from the solenoid.


![](Art/DirectionCurlyEField.png "Magnetic field decreasing and increasing")
**<center> Figure 2. Four varying magnetic fields and rate of change.**


* **<u>Right-hand rule** Point the thumb of your right hand in the direction of the rate of change of the magnetic field ($\frac{-d\vec{B_{1}}}{dt}$) and curl fingers toward your palm. The direction in which the tip of the fingers are pointing is the direction of the curly electric field. 
    
    
* Find the direction of the rate of change of the magnetic field: change $\frac{-d\vec{B_{1}}}{dt}$ to $\frac{-\Delta \vec{B_{1}}}{\Delta t}$, where $\Delta t$ is a small, finite time interval. Then, the direction of $\frac{-\Delta \vec{B_{1}}}{\Delta t}$ = the direction of $-\Delta \vec{B_{1}}$.


* A solenoid with a ring over it will produce a current in the ring. 


    * The resulting $\vec{E_{NC}}$ will be in the same direction, as shown in **Figure 3**. Relate this to a battery, wire, and current.
    

![](Art/MetalRingSolenoid.png "The Effect of a Solenoid on a Metal Ring")
**<center> Figure 3. The Effect of a Solenoid on a Metal Ring**
    
    
* In **Figure 3**, the $I_{2}$ in the ring is proportional to the $\vec{E_{NC}}$ inside the metal. 


* To drive current in a wire, the wire must encircle a region where the magnetic field is changing. If the wire does not, there will be no emf present.

### <center> Faraday's Law
-----------
    
Faraday's Law explains the relationship between $\frac{-\Delta \vec{B_{1}}}{\Delta t}$ and $|\vec{E_{NC}}|$.

#### <center> Magnetic Flux
    
* **Magnitude of the emf**: numerically equal to $|emf| = |\frac{d}{dt}(B_{1}\pi r^{2}_{1})|$


* **Magnetic flux**: calculated in the same way as the electric flux in Chapter 21 on Gauss's law.


    * $(B_{1}\pi r^{2}_{1})$ is the **magnetic flux $\phi_{mag}$** on the area encircled by the circuit.
    
    
    * On a small area: $\delta A$ is $\vec{B}\bullet\hat{n}\delta A = B_{\bot}\delta A$
    
    
        * $\hat{n}$: dimensionless unit vector perpendicular to the area.
        
        
            * "Normal" to the surface.
            
            
        * $B_{\bot}$: the perpendicular component of the magnetic field.
        
        
    * All contributions over the surface are summed to obtain the magnetic flux over the surface.
    
    
    * Can be *positive*, *negative*, or *zero*. 
    
    
        * Depends on the orientation of the magnetic field ($\vec{B}$) relative to $\hat{n}$. 
        
    
#### <center> Faraday's Law Quantitative
    
* **Faraday's Law**: in the presence of a time-varying magnetic field, there is an emf whose magnitude  is equal to the rate of change of magnetic flux.


    * Major physical law concerning the relationship between electric and magnetic fields. 
    
    
    * $emf = -\frac{d \phi_{mag}}{dt} \equiv\oint \vec{E_{NC}} \bullet \vec{dl} = -\frac{d}{dt} \int \vec{B}\bullet\hat{n} dA$
 

    * emf along a round-trip path = rate of change of the magnetic flux on the area encircled by the path.
    
   
    * Direction: Calculate the magnetude of the effect. Then give direction (sign) based on the right-hand rule. 
    

* **Bigger emf**: Faster rate of change of magnetic flux.


    * Also occurs with a bigger area in the same magnetic field.
    
    
#### <center> Coulomb Electric Field in Faraday's Law
    
    
* **Net Electric Field**: $\vec{E} = \vec{E_{C}} + \vec{E_{NC}}$


    * Remember: the round trip integral of a Coulomb electric field is **0**.
    
    
        * Therefore: $emf = \oint \vec{E} \bullet \vec{dl} = \oint (\vec{E_{C}} + \vec_{E_{NC}}) \bullet \vec{dl} = 0 + \oint \vec{E_{NC}} \bullet \vec{dl}$
        
 
#### <center> Voltmeter Readings
    
* **Voltmeter**: acts like an ammeter with a large resistance in series.


* **Voltmeter leads**: when surrounding region of a changing magnetic field, the voltmeter will read an emf equal to $\frac{d\phi_{mag}}{dt}$. 

    
    * If the leads do not surround the changing magnetic field, the voltmeter will read **0 V**. See Figure # for an example.

### <center> Equations
-----------
    
**<center><u>Coulomb Electric Field**
        
\begin{equation*}
\vec{E} = \frac{1}{4\pi\epsilon_{0}}\frac{q}{r^{2}}\hat{r}
\end{equation*}

**<center><u>Force**
        
\begin{equation*}
\vec{F} = q\vec{E}
\end{equation*}

**<center><u>Magnetic Flux**
    
\begin{equation*}
\phi_{mag} = \int\vec{B}\bullet\hat{n}dA= \int B_{\bot} dA
\end{equation*}

**<center><u>Faraday's Law**
        
\begin{equation*}
emf = - \frac{d\phi_{mag}}{dt} \equiv \oint\vec{E_{NC}}\bullet\vec{dl} = -\frac{d}{dt} \int \vec{B}\bullet\hat{n} dA
\end{equation*}

Term | Definition
 :---: | :--- |
