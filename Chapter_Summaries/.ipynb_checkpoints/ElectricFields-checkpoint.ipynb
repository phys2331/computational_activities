### <center> Electric Charge and Force
---------

#### <center> Point Charges

Electric charge can be **positive** or **negative**. 

Like charges (two positive or two negative) **repel** each other. 

Unlike charges (one positive + one negative) **attract** each other.

**Point particle** = object with a very small radius compared to its distance from the other objects of interest. Can treat this point particle as if all of its charge and mass were at a single point. Protons and electrons are almost always considered point particles.

#### <center> The Coulomb Force Law for Point Particles
    
Electromagnetic interaction = one of the four fundamental physical interactions.

**Coulomb's law** = electric force law. Describes magnitude of the electric force between two charged particles.

Equation: 

\begin{equation*}
|{\vec{F}}|=\vec{F}=\frac{1}{4\pi\epsilon_{0}}\frac{|{Q_{1}Q_{2}}|}{r^{2}}
\end{equation*}

$Q_{1}$ and $Q_{2}$ = magnitudes of the electric charge of points 1 and 2, respectively. 
*r* = the distance between the two points.

The electric force acts along a line bewteen the two objects.

#### <center> Units and Constants
    
SI unit of electric charge is the **coulomb (C)**.

Positrons and antiprotons are antimatter.

Positron = antielectron; when interacting with an electron, they will annhilate, releasing their energy as photons.

Ordinary matter is composed of: protons, electrons, and neutrons

#### <center> Size and Structure of Atoms
    
Matter is made of tiny atoms

Solid metal: atoms arranged in a lattice (3D array).

Neutral atom: equal numbers of protons and electrons.

Protons and neutrons are found in the nucleus of the center of the atom.

Electrons are spread in a cloud surrounding the nucleus.

Almost all the mass of an atom is in the nucleus  because the mass of a proton or neutron is ~2000 times larger than that of an electron.

### <center> Concept of the Electric Field
    
---------

The electric field created by a charge is present throughout space at all times.

The electric field penetrates matter. 

definition of electrion field = $\vec{F} = q\vec{E}$

The force on the particle is determined by the charge of the particle and by the electric field it is in from all other particles.

Units of newtons per coulomb (N/C) because $\vec{E} = \frac{\vec{F}}{q}$

If $\vec{E}$ points north, a positive charge will experience a force to the north with the electric field. A negative charge will experience a force to the source, against the electric field. 

#### <center> Drawing the Electric Field
    
Arrows represent the electric field because it is a vector. The tail of the arrow is at the observation locations and the length of the arrow is proportional to the magnitude of the field.

#### <center> Self-Force
    
A point charge is not affected by its own electric field, so it does not have a force exerted upon itself. 

A source charge will make the electric field. A different charge will be affected by that field. So $\vec{F_{on q_{2}}} = q_{2}\vec{E_{1}}$.

#### <center> Physical Concept
    
Field is a physical quantity that has a value at every location in space. Each value can be either scalar or vector.

Example: Temperature in a room is scalar - at each location it has a value. Air flow is a vector field because it moves in a certain direction with a certain speed at each location.

Electric field is a vector because at every location surrounding a charge, the electric field has magnitude and direction.

-----------

### <center> The Electric Field of a Point Charge
    
Equation for the electric field at the observation location:

\begin{equation*}
\vec{E} = \frac{1}{4\pi\epsilon_{0}}\frac{q}{|\vec{r^{2}}|}\hat{r}
\end{equation*}

q = source charge
$\vec{r}$ = position of the observation location relative to the source charge
$\hat{r}$ = unit vector in the direction of $\vec{r}$
$|\vec{r}|$ = magnitude of $\vec{r}$; distance from the source location to the observation location

#### <center> Direction of the Electric Field
    
If the source charge (q) = positive, $q\hat{r}$ is in the direction of $\hat{r}$, away from the source charge. 

If the source charge (q) = negative, $q\hat{r}$ is in the direction of $-\hat{r}$, toward the source charge. 

#### <center> Distance dependence of electric field
    
The electric field of a point charge depends strongly on distance. Shown in equation because it uses the square of the distance in the denominator. 

Example: Double the distance between the source location and observation location, the denominator gets 4 times bigger, meaning the electric field is 1/4 what it was before.

#### <center> Magnitude of charge
    
The larger the charge of the source particle, the larger the magnitude of the electric field (because charge is in the numerator).

#### <center> Magitude of Electric Field
    
$\frac{1}{4\pi\epsilon_{0}}\frac{q}{|\vec{r^{2}}|}$ is a scalar quantity that doesn't equal the magnitude of the electric field because the charges can be positive or negative and vector magnitudes can be only positive. Take the absolute value of this equation to find the magnitude of the electric field. 

#### <center> Patterns of Electric Field Near Point Charges
    
Electric field of a point charge is spherically symmetric. 

The magnitude of the electric field at each particular distance (r) away is the same.

The magnitude of the field decreases with distance from the charge.

The electric field of a negative source charge points inward and the magnitude decreases with increased distance from the charge.

#### <center> Electric field of a Uniformly Charged Sphere
  
A spherical object with a charge uniformaly spread over the surface creates an electric field. The electric field due to the sphere is the same as if all its charge were in the center instead of the surface. 

Inside the sphere, the electric field is 0 because it's the sum of all charges on the surface. 

Can use a uniformly charged sphere as a point charge because the E field is the same as if there were a point charge at the center of the sphere and because it responds to other E fields in the same way a point charge would.

### <center> Superposition of Electric Fields
    
Superposition principle for electric forces: the net electric field due to two or more charges is the vector sum of each field due to each individual charge. 

Matter cannot block electric fields. The electric field due to a charged particle is unaffected by the presence of other charged particles.
