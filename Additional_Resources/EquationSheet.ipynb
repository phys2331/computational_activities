## <center> Physics II Equation Sheet

____


### <center> Force
    
___
    
The force felt by a particle with charge q moving at velocity $\vec{v}$:

Due to an **electric field $\vec{E}$**:
    
\begin{equation*}
\vec{F}_{elec} = q\vec{E}
\end{equation*}

Due to a **magnetic field $\vec{B}$**:

\begin{equation*}
\vec{F}_{mag} = q\vec{v}\times\vec{B}
\end{equation*}

$\vec{F} = m\vec{a}$

___

### <center> Electric Field
    
___

#### <center> $\vec{E}$ created by point charge (q) at a location $\vec{r}$ away
    
\begin{equation*}
\vec{E} (\vec{r}) = \frac{kq}{|\vec{r}|^{2}} \hat{r}
\end{equation*}

#### <center> Net $\vec{E}$ due to multiple charges

\begin{equation*}
\vec{E}_{net} = \frac{kq_{1}}{|\vec{r}_{1}|^{2}} \hat{r}_{1} + \frac{kq_{2}}{|\vec{r}_{2}|^{2}} \hat{r}_{2} + ...
\end{equation*}

#### <center> Curly $\vec{E}$ due to accelerating charge

\begin{equation*}
\oint \vec{E} \bullet \vec{dl} = -\frac{d}{dt} \int \vec{B} \bullet \vec{dA}
\end{equation*}

___

### <center> Magnetic Field
    
___
    
\begin{equation*}
\vec{dB} = \frac{\mu_{0}}{4\pi} \frac{I\vec{dl}\times\hat{r}}{|\vec{r}|^{2}}
\end{equation*}

\begin{equation*}
\vec{B} = \vec{dB_{1}} + \vec{dB_{2}} + ...
\end{equation*}

\begin{equation*}
\oint \vec{B} \bullet \vec{dl} = \epsilon_{0} \frac{d}{dt} \int \vec{E} \bullet \vec{dA}
\end{equation*}

___

### <center> Voltage
    
___

\begin{equation*}
\Delta V = - \int_{A}^{B} \vec{E} \bullet \vec{dl}
\end{equation*}

\begin{equation*}
\Delta V = \frac{\Delta U}{q}
\end{equation*}

___

### <center> Conservation of Energy
    
___

\begin{equation*}
\Delta KE + \Delta U = 0
\end{equation*}

<center> where
    
\begin{equation*}
\Delta U = -W = - \int_{A}^{B} \vec{F} \bullet \vec{dl}
\end{equation*}

\begin{equation*}
KE = \frac{1}{2}mv^{2}
\end{equation*}
